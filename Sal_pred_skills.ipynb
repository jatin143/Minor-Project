```python
import re
```

```python
def clean_skills(skl):
    skills = str(skl).lower()
    skills = re.sub('\...','',skills)
    skills = re.sub(',','',skills)
    skills = re.sub(r'\s+', ' ', skills)
    return skills

df['Key_skills'] = df['key_skills'].apply(clean_skills)
```

```python
df.head()
```

```python
df.job_description.fillna('missing',inplace = True)
```

```python
def clean_job_desc(job):
    job_desc = str(job).lower()
    job_desc = re.sub(r'[^a-z]', ' ', job_desc)
    job_desc = re.sub(r'\s+', ' ', job_desc)
    return job_desc

df['job_desc'] = df['job_description'].apply(clean_job_desc)
```

Transforming the Jobs column with consistent values having same meaning

```python
df['job_type'].fillna('missingjobtype', inplace=True)
df['job_type'].replace('Analytics', 'analytics', inplace=True)
df['job_type'].replace('Analytic', 'analytics', inplace=True)
df['job_type'].replace('ANALYTICS', 'analytics', inplace=True)
df['job_type'].replace('analytic', 'analytics', inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
def min_exp(val):
    exp = re.sub('-',' ',val)
    exp = exp.split(" ")
    exp = int(exp[0])
    return exp
    
def max_exp(val):
    exp = re.sub('-',' ',val)
    exp = exp.split(' ')
    exp = int(exp[1])
    return exp
```

```python
df['min_exp'] = df['experience'].apply(lambda x : min_exp(x))
df['max_exp'] = df['experience'].apply(lambda x : max_exp(x))
```

```python
def clean_job_desig(desig):
    job_desig = desig.lower()
    job_desig = re.sub(r'[^a-z]', ' ', job_desig)
    job_desig = re.sub(r'\s+', ' ', job_desig)
    return job_desig
```

```python
df['Job_desig'] = df['job_desig'].apply(clean_job_desig)
```

```python
df['Emp_info'] = (df['Job_desig'] + ' ' + df['job_desc'] + ' ' + df['job_type'])
```

```python
df.head()
```

```python
df1=df[['min_exp','max_exp','Emp_info','Key_skills','salary']]
```
