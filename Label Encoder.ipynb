## Using Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder()
label=l.fit_transform(df["Designation"])
```

```python
labels=l.fit_transform(df["Department"])
```

```python
labels
```

```python
label
```

```python
df.drop(columns=['Designation','Date_of_Birth','Name','Age','Join_Date','GROSS','Deduction'],inplace=True)
```

```python
df['Designation']=label
df
```

```python
df.drop(columns=['Department'],inplace=True)
```

```python
df['Department']=labels
df
```

```python
dumm = pd.get_dummies(df.Gender)
dumm
```

```python
df4=pd.concat([df,dumm],axis='columns')
df4
```

```python
df4.drop(columns=['Gender'],inplace=True)
```

```python
df4
```

```python
a=df4.drop('Net_Pay',axis='columns')
a
```

```python
b=df4.Net_Pay
b
```

```python
from sklearn.linear_model import LinearRegression
mod = LinearRegression()
```

```python
mod.fit(a,b)
```

```python
mod.predict(a)
```

```python
mod.score(a,b)
```
